Joint MSSEG2-Inverse



```python
list_txt = glob.glob("/content/res_anima/MSSEG-15-5/seqjoint-inv/last*15_5*.txt")
list_txt
```

```python
df_joint = pd.read_csv('/content/MSSEG-2_whole_inv_recalc_csf_lastItr.csv', sep=';')
df_joint
df_joint = df_joint.drop_duplicates()
df_joint = df_joint[(df_joint.Lambda2 == 15) & (df_joint.Lambda3 == 5)]
df_joint.Algorithm = "joint"
df_joint.OriginDataset = "MSSEG"
df_joint["sens"] = "inverse"
df_joint
```

```python
new_df = pd.DataFrame(data = None, columns = ['Jaccard','Dice', 
                            'Sensitivity',
                            'Specificity',
                            'PPV',
                            'NPV',
                            'RelativeVolumeError',
                            'HausdorffDistance',
                            'ContourMeanDistance',
                            'SurfaceDistance',
                            'PPVL',
                            'SensL',
                            'F1_score','patientId','Lambda1','Lambda2', 'Lambda3'])
for f in list_txt:
  basename = os.path.basename(f)
  npatient = basename.split('_')[-10]
  lambda2 = basename.split('_')[-4]
  lambda3 = 3
  lambda1 = 0
  df0 = pd.read_csv(f, sep=';', \
                    names=['Jaccard','Dice', 
                            'Sensitivity',
                            'Specificity',
                            'PPV',
                            'NPV',
                            'RelativeVolumeError',
                            'HausdorffDistance',
                            'ContourMeanDistance',
                            'SurfaceDistance',
                            'PPVL',
                            'SensL',
                            'F1_score','patientId','Lambda1','Lambda2', 'Lambda3'])
  df0.Lambda1 = 70
  df0.Lambda2 = lambda2
  df0.Lambda3 = 3
  df0.patientId = npatient
  new_df = new_df.append(df0, ignore_index=True)

new_df
```

```python
new_df = new_df.replace({'\t': ' '}, regex=True)
new_df = new_df.replace(r'^\s*$', np.nan, regex=True)
new_df = new_df.convert_dtypes()
new_df = new_df.astype({"Lambda1" : np.int64, "Lambda2" : np.float64, "Lambda3" : np.float64, "patientId" : np.int64})
print(new_df.dtypes)
```

```python
print(df_joint.dtypes)
```

```python
jointure = df_joint.copy()
jointure = jointure.merge(new_df, on=['patientId', 'Lambda2', 'Lambda3'], suffixes=['', '_anima'] )
jointure.OriginDataset = 'MSSEG'
jointure
```

```python
jointure.to_csv('joint_MSSEG_15_3_inverse_anima.csv', index=False,sep=';')  
```
